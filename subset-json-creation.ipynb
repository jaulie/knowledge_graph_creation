```python
%cd "/Users/rebeccaglick/Desktop/pubmedqa/data"

import json

# Load data
with open('ori_pqal.json', 'r') as f:
    data = json.load(f)

# List of diabetes/endocrine-related IDs
diabetes_ids = [
    '15703931', '27456836', '22720085', '26556589', '24139705', '15939071', '18948835',
    '15125825', '12006913', '23224030', '26536001', '22266735', '17894828', '12145243',
    '16971978', '22522271', '15095519', '16241924', '15787677', '19406119', '8521557',
    '8738894', '20011163'
]

# Extract matching entries
filtered_data = {id_: data[id_] for id_ in diabetes_ids if id_ in data}

# Optional: print or inspect a few
for id_, entry in filtered_data.items():
    print(f"\n=== ID: {id_} ===")
    print(f"Question: {entry.get('QUESTION')}")
    print(f"Year: {entry.get('YEAR')}")
    print(f"MeSH: {entry.get('MESHES')}")
    
    print(f"Answer: {entry.get('LONG_ANSWER')}")
    print("-" * 60)

# Optional: save to new JSON file
with open('diabetes_subset.json', 'w') as f:
    json.dump(filtered_data, f, indent=2)
```
